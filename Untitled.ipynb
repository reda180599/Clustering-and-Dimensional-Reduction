# PCA and Dimensionality Reduction

```python
#Toolbox

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
data = pd.read_csv('data.csv')

data.head()
```

```python
data.drop(['Unnamed: 32', 'id'],axis=1, inplace=True)
```

### ModelBuildng - Divide Data into Train and Test


```python

X = data.drop(['diagnosis'], axis=1)
y = data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,random_state=0, test_size=0.2)
```

```python
#convert object to int (diagnosis)
#data['diagnosis'] = pd.get_dummies(data, columns=['diagnosis'], drop_first=True)
```

```python
#data.info()
```

### Scalling

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

### Model Building

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_train_scaled, y_train)

y_predict = lr.predict(x_test_scaled)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predict))
```

```python
#print(data.columns.value_counts().sum()) # number of columns / features
```

### Diminsionality Reduction - Extract new fatures from existed features

```python
from sklearn.decomposition import PCA # PCA == Principal Component Analysis

pca = PCA(n_components=4)

x_train_pca = pca.fit_transform(x_train_scaled)
x_test_pca = pca.transform(x_test_scaled)
```

```python
x_train_pca
```

```python
sns.scatterplot(x_train_pca[:, 0], x_train_pca[:, 1], hue=y_train)
```

```python
lr.fit(x_train_pca, y_train)

y_predict_pca = lr.predict(x_test_pca)
```

```python
print(classification_report(y_test, y_predict_pca))
```

#### When we have 31 features we get 97% accuracy and when we use Diminsionality Reduction and minimize features to 4 we get 96%accuracy

### Using SVM model (non linear data)

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear') # to make it linear

svm.fit(x_train_pca, y_train)

y_predict_svm = svm.predict(x_test_pca)
```

```python
print(classification_report(y_test, y_predict_svm))
```
